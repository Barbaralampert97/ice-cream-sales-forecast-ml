### Exploração dos dados (EDA)
Este notebook tem como objetivo explorar o dataset `Ice Cream.csv` para compreender a relação entre temperatura ambiente e faturamento de sorvetes. 
Serão realizadas análises estatísticas e gráficos para visualizar padrões e possíveis tendências.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_df = pd.read_csv('../data/Ice Cream.csv')
```

```python
data_df.head()
```

```python
print("Número de linhas e colunas: ", data_df.shape)
```

```python
print('Informações do DataFrame:', data_df.info())
```

```python
print('Resumo estatístico do DataFrame:\n',data_df.describe())
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data_df['Temperature'], data_df['Revenue'], color='red')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Faturamento')
plt.title('Temperatura vs Faturamento de Sorvetes')
plt.grid(True)
plt.show()
```

```python
plt.hist(data_df['Temperature'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Frequência')
plt.title('Distribuição da Temperatura')
plt.show()

plt.hist(data_df['Revenue'], bins=10, color='orange', edgecolor='black')
plt.xlabel('Faturamento')
plt.ylabel('Frequência')
plt.title('Distribuição do Faturamento')
plt.show()
```

```python
# Matriz de correlação
print('Matriz de correlação:\n', data_df.corr())

# Visualizar correlação
plt.figure(figsize=(6,4))
plt.imshow(data_df.corr(), cmap='coolwarm', aspect='auto')
plt.colorbar()
plt.title('Matriz de Correlação')
plt.xticks([0, 1], data_df.columns)
plt.yticks([0, 1], data_df.columns)
plt.show()
```

### 📊 Insights da Análise Exploratória

Após analisar o dataset de vendas de sorvetes, identificamos os seguintes insights:

**1. Forte Correlação Positiva**
- Existe uma relação linear clara entre temperatura e faturamento
- A matriz de correlação confirma uma correlação forte (próxima de 1)
- Conclusão: quanto maior a temperatura, maior o faturamento

**2. Distribuição dos Dados**
- A temperatura apresenta uma distribuição que indica padrões sazonais
- O faturamento acompanha a distribuição da temperatura, reforçando a correlação
- Não foram identificados outliers significativos

**3. Qualidade dos Dados**
- Dataset completo, sem valores nulos
- Todas as variáveis estão no formato numérico correto
- Os dados estão prontos para modelagem
